**1- Console-Based Smart Calculator**

```python
import numpy as n #importing numpy as n to store elements in an array
#Displaying a menu with operations
print("Arithmetic Operations calculator")
print("1-Addition")
print("2-Subtraction")
print("3-Multiplication")
print("4-Division")
print("5-Power calculation")
print("6-Modulus Operation")
choice=int(input("Enter the number from the above choices for applying operation:")) #to know whicch operation is going to be performed
num=int(input("Enter how many numbers?:")) #on how many numbers user want to operate
list_1=[] #creating an empty list to store the elements
for i in range(num):
  a=int(input(f"Enter number {i+1}:"))
  list_1.append(a)
array_1=n.array(list_1) #array is also a list so we can  store list in numpy array
if num>1: #if there is more than 1 operand available only then calculator will work that's why usinf if-else here
#choices are already given above
  if choice==1:
    add=n.sum(array_1)
    print(f"Addition of numbers : {add}")
  elif choice==2:
    sub=array_1[0]
    for i in range(1,num):
      sub=sub-array_1[i]
    print(f"Subtraction of numbers : {sub}")
  elif choice==3:
    mul=1
    for i in range(0,num):
      mul=mul*array_1[i]
    print(f"Multiplication of numbers : {mul}")
  elif choice==4:
    if array_1[0]==0: #denominator should not be zero!
      print("Division by zero is not possible")
    else:
      div=array_1[0]
      for i in range(1,num):
        div=div/array_1[i]
      print(f"Division of numbers : {div}")
  elif choice==5:
    power=array_1[0]
    for i in range(1,num):
      power=power**array_1[i]
    print(f"Power of numbers : {power}")
  elif choice==6:
    if array_1[0]==0:  #denominator should not be zero!
      print("Division by zero is not possible")
    else:
      mod=array_1[0]
      for i in range(1,num):
        mod=mod%array_1[i]
      print(f"Modulus of numbers : {mod}")
  else:
    print("Invalid choice. Try Again!!")
else:
  print("Oopss!Invalid input...Try again and put numbers more than 1")
```

**2-Student Marks Analyzer**

```python
import numpy as np #Import numpy as np to store the marks of student in array
student=int(input("Enter the number of students: "))
list_2=[]
for i in range(student):
  marks=int(input(f"Enter the marks(less than 100) of student{i+1}: "))
  list_2.append(marks)
array_2=np.array(list_2)
if student>1:
  if marks<=100:
    avg=np.mean(array_2)
    print(f"Total marks:{np.sum(array_2)}")
    print(f"Average marks:{avg}")
    print(f"Highest marks:{np.max(array_2)}")
    print(f"Lowest marks:{np.min(array_2)}")
    if avg>=85:
      print("Grade:A")
    elif avg>=70 and avg<85:
      print("Grade:B")
    elif avg>=50 and avg<70:
      print("Grade:C")
    else:
      print("Fail")
  else:
    print("Invalid input. Marks should be less than 100")
else:
  print("Oopss!Invalid input...Try again and give input for student6 more than 1")
```
